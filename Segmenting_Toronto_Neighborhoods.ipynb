## Code to Scrape data from Wiki Link

```python
# Empty Lists for column values.
postcodes = []
boroughs = []
neighborhoods = []
```

```python
# Accessing source code of wiki link with Toronto neighborhood data
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'html')
print(soup.prettify())
```

```python
# Saves table entries in each row into the column lists
web_data_table = soup.find('table')
table_body = web_data_table.find('tbody')
for row in table_body.find_all('tr'):
    col_counter = 0
    for entry in row.find_all('td'):
        if entry == []:
            continue
        value = entry.text
        if col_counter == 0:
            postcodes.append(value)
        elif col_counter == 1:
            boroughs.append(value)
        else:
            neighborhoods.append(value.split('\n')[0])
        print(value)
        col_counter = col_counter + 1
```

## Code to maniputlate scraped data into propper Data format

```python
# Saves column lists to the Neighborhood Data Frame
neighborhood_df['PostalCode'] = postcodes
neighborhood_df['Borough'] = boroughs
neighborhood_df['Neighborhood'] = neighborhoods
print(neighborhood_df.head())
print(neighborhood_df.shape)
```

```python
# Removes row entries with no assigned boroughs
trimmed_neighborhood_df = neighborhood_df[neighborhood_df['Borough'] != 'Not assigned']
trimmed_neighborhood = trimmed_neighborhood_df.reset_index(drop=True)
print(trimmed_neighborhood_df.head())
print(trimmed_neighborhood_df.shape)
```

```python
# Captures the row entries that have Neighborhoods with no assigned value
index_to_clean = trimmed_neighborhood_df[trimmed_neighborhood_df['Neighborhood'] == 'Not assigned'].index
print(index_to_clean)
```

```python
# Resests not assigned Neighborhood values to Borough values
trimmed_neighborhood_df.loc[index_to_clean, 'Neighborhood'] = trimmed_neighborhood_df.loc[index_to_clean, 'Borough'] 
```

```python
print(trimmed_neighborhood_df.head(10))
print(trimmed_neighborhood_df.shape)
```

```python
# Aggregates common postal code and Borough values into one row while perserving all uniqure neighborhood values
final_neighborhood_df = trimmed_neighborhood_df.groupby('PostalCode').agg({'PostalCode':'first', 
                             'Borough': 'first', 'Neighborhood': ', '.join}).reset_index(drop=True)
print(final_neighborhood_df.head(20))
print(final_neighborhood_df.shape)
```

### Clustering Analysis for Boroughs with Toronto in the name.

```python
#function to look up venues around each neighborhood.
def getNearbyVenues(names, latitudes, longitudes):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#Grabbing the venues around the Toronto Boroughs
toronto_venues = getNearbyVenues(names=toronto_neighborhoods['Neighborhood'],
                                   latitudes=toronto_neighborhoods['Latitude'],
                                   longitudes=toronto_neighborhoods['Longitude']
                                  )
```

Creating Dataframe to analyze each Neighborhood

```python
toronto_onehot =[]

# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood Name'] = toronto_venues['Neighborhood'] 

#Neighborhood Name column index
col_ind = toronto_onehot.columns.get_loc("Neighborhood Name")

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[col_ind]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood Name').mean().reset_index()
toronto_grouped
```

Checking the size of the dataframe created.

```python
toronto_grouped.shape
```

Printing the name of each Neighborhood name and the top 10 categories in each.

```python
num_top_venues = 10

for hood in toronto_grouped['Neighborhood Name']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood Name'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Putting what was printted in the cells above to a pandas dataframe.

```python
#function to sort the venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood Name']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood Name'] = toronto_grouped['Neighborhood Name']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### Cluster Neighborhood

Run *k*-means to cluster the neighborhood into 5 clusters.

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood Name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_neighborhoods

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood Name'), on='Neighborhood')

toronto_merged.head() # check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
